```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = pd.read_csv('crimes_against_women.csv')
```

```python
print(dataset)
```

```python
dataframe = []

for row in dataset.values:
    if('TOTAL' in row[2].upper()):
        dataframe += [[row[1], row[4], row[5], row[6], row[7], row[8], row[9], row[10]]]
```

```python
dataframe = pd.DataFrame(dataframe, columns=['STATE/UT', 'RAPE', 'KIDNAPPING', 'DOWRY', 'ASSAULT', 'INSULT', 'CRUELTY', 'TRAFFECKING'])
dataframe.iloc[:, 0] = dataframe.iloc[:, 0].str.upper()
```

```python
X = dataframe.iloc[:, 0].values
y = dataframe.iloc[:, 1].values
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X = le.fit_transform(X)
```

```python
plt.bar(X, y)
plt.xlabel('States')
plt.ylabel('Rapes')
plt.title('Unfortunate Graph')
plt.plot()
```

```python
arr = []
print(le.inverse_transform([0])[0])
for i in range(0, 36):
    arr += [(i, (le.inverse_transform([i]))[0])]
print(arr)
```
